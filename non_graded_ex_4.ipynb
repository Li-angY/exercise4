```python
import random
import time

# Define the cards with their characteristics using a dictionary
card_deck = [
    {"Name": "Diablo", "Health": 100, "Attack": 90, "Defense": 60},
    {"Name": "Medusa", "Health": 100, "Attack": 70, "Defense": 70},
    {"Name": "Jester", "Health": 120, "Attack": 60, "Defense": 90},
    {"Name": "Troll", "Health": 150, "Attack": 40, "Defense": 94},
    {"Name": "Specter", "Health": 100, "Attack": 70, "Defense": 70},
    {"Name": "Mist", "Health": 100, "Attack": 75, "Defense": 65},
    {"Name": "Savage", "Health": 100, "Attack": 90, "Defense": 50},
    {"Name": "Marauder", "Health": 100, "Attack": 85, "Defense": 50},
    {"Name": "Wimp", "Health": 110, "Attack": 40, "Defense": 85},
    {"Name": "Sorcerer", "Health": 100, "Attack": 70, "Defense": 55}
]

# Number of cards to be dealt to each player
deck_size = 5

# Draw random cards for player and opponent
player_deck = random.sample(card_deck, deck_size)
opponent_deck = random.sample([card for card in card_deck if card not in player_deck], deck_size)

# Function to display card details
def display_card(card):
    print(f"Name: {card['Name']}, Health: {card['Health']}, Attack: {card['Attack']}, Defense: {card['Defense']}")

# Function to handle the battle mechanics between two cards
def battle(attacker, defender):
    print(f"\n{attacker['Name']} attacks {defender['Name']}!")
    if attacker['Attack'] > defender['Health']:
        print(f"{defender['Name']} is defeated instantly!")
        return True  # Defender is defeated
    else:
        damage_to_defense = min(attacker['Attack'], defender['Defense'])
        defender['Defense'] -= damage_to_defense
        remaining_attack = attacker['Attack'] - damage_to_defense
        if remaining_attack > 0:
            defender['Health'] -= remaining_attack
        print(f"Damage dealt to {defender['Name']} - Defense reduced by {damage_to_defense}, Health reduced by {remaining_attack}.")
        display_card(defender)
        return defender['Health'] <= 0  # Return True if defender is defeated

# Main game loop
while player_deck and opponent_deck:
    # Display player's deck
    print("\nYour deck:")
    for idx, card in enumerate(player_deck):
        print(f"{idx + 1}. {card['Name']}")

    # Opponent draws a random card
    opponent_card = random.choice(opponent_deck)
    print("\nOpponent has drawn a card:")
    display_card(opponent_card)

    # Player chooses a card to fight
    while True:
        try:
            choice = int(input("\nChoose a card number to fight (1-{}): ".format(len(player_deck)))) - 1
            if 0 <= choice < len(player_deck):
                player_card = player_deck[choice]
                break
            else:
                print("Invalid choice. Please choose a valid card number.")
        except ValueError:
            print("Please enter a valid number.")

    # Execute battle
    if battle(player_card, opponent_card):
        opponent_deck.remove(opponent_card)

    # Check if the opponent still has cards
    if not opponent_deck:
        print("\nOpponent has no more cards. You win!")
        break

    # Pause before the opponent's turn
    time.sleep(3)

    # Opponent's turn
    print("\nOpponent's turn...")

    # Draw a random card from opponent deck
    opponent_card = random.choice(opponent_deck)
    print("\nOpponent chose a card:")
    display_card(opponent_card)

    # Display player's available cards
    print("\nYour deck:")
    for idx, card in enumerate(player_deck):
        print(f"{idx + 1}. {card['Name']}")

    # Player chooses a card for defense
    while True:
        try:
            choice = int(input("\nChoose a card number to defend (1-{}): ".format(len(player_deck)))) - 1
            if 0 <= choice < len(player_deck):
                player_card = player_deck[choice]
                break
            else:
                print("Invalid choice. Please choose a valid card number.")
        except ValueError:
            print("Please enter a valid number.")

    # Opponent's battle logic (mirrors player's logic)
    if battle(opponent_card, player_card):
        player_deck.remove(player_card)

    # Check if the player still has cards
    if not player_deck:
        print("\nYou have no more cards left. The opponent wins!")
        break

    # Pause before the next round
    time.sleep(3)

# Check game outcome
if not player_deck:
    print("\nYou have no more cards left. The opponent wins!")
elif not opponent_deck:
    print("\nOpponent has no more cards left. You win!")
```
